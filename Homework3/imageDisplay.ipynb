```python
from matplotlib import pyplot as plt
import numpy as np

dir = "output/run23"

dataIn = np.load(dir+"/sampleIn.npy")
dataOut = np.load(dir+"/sampleOut.npy")

for i in range(dataIn.shape[0]):
    fig=plt.figure(figsize=(5, 5))
    fig.add_subplot(1, 2, 1)
    plt.imshow(dataIn[i], interpolation='nearest')
    fig.add_subplot(1, 2, 2)
    plt.imshow(dataOut[i], interpolation='nearest')
    plt.show()
```
